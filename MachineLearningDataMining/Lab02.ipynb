#### Exercise 4: Normalization

````python
import pandas as pd

df = pd.read_csv('data.csv')
df
````

##### __(a) Implement a python function, which uses  linear min-max normalization to transform the column “Size”.__

````python
import numpy as np
sizeColumn = df['Size (cm)']
normalizedlin_DF = (sizeColumn - sizeColumn.min()) / ((sizeColumn.max())-sizeColumn.min())
normalizedlin_DF
````
